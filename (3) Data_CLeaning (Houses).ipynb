```python
import re
df['society'] = df['society'].apply(
    lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()
).str.lower()
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100, 2)
        else:
            return round(float(x[0]), 2)
        


# def treat_price(x):
#     if isinstance(x, float) or isinstance(x, int):
#         return round(x, 2)
#     elif isinstance(x, (list, tuple)) and len(x) == 2:
#         amount, unit = x
#         try:
#             amount = float(amount)
#             if unit.lower() == 'lac':
#                 return round(amount / 100, 2)  # Convert Lac to Crores
#             elif unit.lower() == 'crore':
#                 return round(amount, 2)
#             else:
#                 raise ValueError(f"Unknown unit: {unit}")
#         except ValueError:
#             return None  # or handle error as needed
#     else:
#         return None  # or handle unexpected format
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.rename(columns={'rate': 'price_per_sqft'}, inplace=True)
df.head(2)
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace(',', '').str.replace('₹', '').astype(float)
```

```python
df.head(2)
```

```python
df['area'] = ((df['price']*10000000)/df['price_per_sqft']).round()
```
